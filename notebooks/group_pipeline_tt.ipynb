```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
```

```python
import nltk
nltk.download('stopwords')

```

```python
# Step 2: Load Dataset
# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

#Adjust the path
file_path = '/content/drive/MyDrive/AIML assignment/data/raw/IMDB_Dataset.csv'
df = pd.read_csv(file_path)
```

```python
# Step 3: Data Cleaning
def clean_text(text):
    text = re.sub('<[^>]*>', '', text) #Remove HTML tags
    text = re.sub('[^a-zA-Z]', ' ', text).lower() #Remove punctuation and lowercase
    return text

df['review'] = df['review'].apply(clean_text)
```

```python
# Step 4: Stop Words Removal
stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    words = text.split()
    words = [w for w in words if not w in stop_words]
    return ' '.join(words)

df['review'] = df['review'].apply(remove_stopwords)

```

```python
# Step 5: Stemming
stemmer = PorterStemmer()

def stem_text(text):
    words = text.split()
    words = [stemmer.stem(w) for w in words]
    return ' '.join(words)

df['review'] = df['review'].apply(stem_text)
```

```python
# Step 6: Lowercasing (Handled in Data Cleaning)
# Step 7: Encoding Categorical Variables
encoder = LabelEncoder()
df['sentiment'] = encoder.fit_transform(df['sentiment'])

# Display Processed Data
print(df.head())

# Step 8: Create processed_data directory
import os
output_dir = '/content/drive/MyDrive/AIML assignment/results'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
```

```python
# Step 9: Save the preprocessed data to CSV
output_file_path = os.path.join(output_dir, 'preprocessed_reviews.csv')
df.to_csv(output_file_path, index=False)

print(f"Preprocessed data saved to {output_file_path}")
```
